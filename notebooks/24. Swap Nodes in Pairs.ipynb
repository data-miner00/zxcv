# Swap Nodes in Pairs

Difficulty: Medium

Given a linked list, swap every two adjacent nodes and return its head. You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)

## Examples 

Example 1:

    Input: head = [1,2,3,4]
    
    Output: [2,1,4,3]

Example 2:

    Input: head = []
    
    Output: []

Example 3:

    Input: head = [1]
    
    Output: [1]

Example 4:

    Input: head = [1,2,3]
    
    Output: [2,1,3]

## Constraints

- The number of nodes in the list is in the range [0, 100].
- 0 <= Node.val <= 100

## Double Pass

### Solution 1

1. Have two placeholder list
2. Assign to each of them in an alternate manner
3. Second loop matches back the changes starting from second list.

Time complexity: $O(n)$

Submission link: https://leetcode.com/problems/swap-nodes-in-pairs/submissions/1782317265/

```python
from typing import Optional, List

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        first = None
        first_ptr = None
        second = None
        second_ptr = None

        current = head
        index = 0
        while current:
            if index % 2 == 0:
                if not second and not second_ptr:
                    second = current
                    second_ptr = current
                else:
                    second_ptr.next = current
                    second_ptr = second_ptr.next

                if not current.next and first_ptr:
                    first_ptr.next = None
            else:
                if not first and not first_ptr:
                    first = current
                    first_ptr = current
                else:
                    first_ptr.next = current
                    first_ptr = first_ptr.next

                if not current.next and second_ptr:
                    second_ptr.next = None
            
            index += 1
            current = current.next

        if first and not second:
            return first
        if second and not first:
            return second

        final = None
        final_ptr = None

        index = 0
        
        while first and second:
            if not final and not final_ptr:
                final = first
                final_ptr = first
                first = first.next
            else:
                if index % 2 == 0:
                    final_ptr.next = second
                    second = second.next
                else:
                    final_ptr.next = first
                    first = first.next
                final_ptr = final_ptr.next
                index += 1

        if first:
            final_ptr.next = first
        if second:
            final_ptr.next = second

        return final
```

## Test cases

```python
sln = Solution()
```

```python
def makeList(nums: List[int]) -> Optional[ListNode]:
    if not nums: return None
    head = None
    pointer = None
    
    for i in nums:
        current = ListNode(i)
        if not head:
            head = current
            pointer = current
        else:
            pointer.next = current
            pointer = pointer.next

    return head
```

```python
def collectToList(nums: Optional[ListNode]) -> List[int]:
    if not nums: return []
    
    node = nums
    res = [node.val]
    while node.next:
        node = node.next
        res.append(node.val)

    return res
```

```python
scenarios = [
    [makeList([1,2,3,4]), [2,1,4,3]],
    [makeList([]), []],
    [makeList([1]), [1]],
    [makeList([1,2,3]), [2,1,3]],
    [makeList([1,2]), [2,1]],
]

for case in scenarios:
    actual = collectToList(sln.swapPairs(case[0]))
    print('Actual   : ', actual)
    print('Expected : ', case[1])
    print('-' * 50)
    assert actual == case[1], f"Case '{case[0]}' failed. {actual} does not equal to {case[1]}"
```
