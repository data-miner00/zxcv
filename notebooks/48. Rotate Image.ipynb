# 48. Rotate Image

Difficulty: Medium

You are given an n x n 2D matrix representing an image, rotate the image by 90 degrees (clockwise).

You have to rotate the image in-place, which means you have to modify the input 2D matrix directly. DO NOT allocate another 2D matrix and do the rotation.

- 1 <= n <= 20
- -1000 <= matrix[i][j] <= 1000

## Transpose

### Solution 1 (Claude)

Step 1 - Transpose:

- Swap matrix[i][j] with matrix[j][i]
- Loop j from i+1 to avoid swapping twice
- This converts rows to columns

Step 2 - Reverse:

- Reverse each row individually
- This completes the 90Â° clockwise rotation

Submission link: https://leetcode.com/problems/rotate-image/submissions/1821478236/

```python
from typing import List

class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        n = len(matrix)
    
        # Step 1: Transpose the matrix
        for i in range(n):
            for j in range(i + 1, n):  # Only traverse upper triangle
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
        
        # Step 2: Reverse each row
        for i in range(n):
            matrix[i].reverse()
        
```
